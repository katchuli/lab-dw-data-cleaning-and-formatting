## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
def normalize_gender(x):
    F_target = ["female", "Femal"]
    M_target = ["Male"]
    
    if x in F_target:
        return "F"
    elif x in M_target: 
        return "M"
    else: 
        return x

file["gender"] = list(map(normalize_gender, file["gender"]))
display(file["gender"].value_counts())
```

```python
# State names:
print(file["state"].unique())
```

```python
def state_name(x):
    WA_target = ["Washington", "WA"]
    AZ_target = ["Arizona", "AZ"]
    CF_target = ["California", "Cali"]
    
    if x in WA_target:
        return "Washington"
    elif x in AZ_target: 
        return "Arizona"
    elif x in CF_target:
        return "California"
    else: 
        return x

file["state"] = list(map(state_name, file["state"]))
display(file["state"].value_counts())
    
```

```python
#In education, "Bachelors" could be replaced by "Bachelor"
file['education'] = file['education'].str.replace('Bachelors', 'Bachelor')
```

```python
file["education"] = list(map(state_name, file["education"]))
display(file["education"].value_counts())
```

```python
#In Customer Lifetime Value, delete the % character
file['customer_lifetime_value'] = file['customer_lifetime_value'].str.replace('%', ' ')
file
```

```python
# In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"
file['vehicle_class'] = file['vehicle_class'].str.replace('Sports Car', 'Luxury')
```

```python
file['vehicle_class'] = file['vehicle_class'].str.replace('Luxury SUV', 'Luxury')
```

```python
file['vehicle_class'] = file['vehicle_class'].str.replace('Luxury Car', 'Luxury')
file.head(40)
file.tail(40)
```

Number of open complaints is a string - remember you can use `split()` to deal with it and take the number you need. Finally, since it should be numeric, cast the column to be in its proper type.*

```python
file.dtypes
```

```python
file['customer_lifetime_value'] = pd.to_numeric(file["customer_lifetime_value"], errors='coerce')
file.dtypes
```

```python
unique_values = file['number_of_open_complaints'].unique()
print(unique_values)
```

```python
def middle_value(complaints):
    if pd.isna(complaints):
        return complaints  # Return NaN as is
    parts = str(complaints).split('/')
    if len(parts) >= 2:
        return parts[1]
    else:
        return complaints  # Return original value if split doesn't yield at least two parts

file['number_of_open_complaints'] = file['number_of_open_complaints'].apply(middle_value)
file['number_of_open_complaints'] = pd.to_numeric(file['number_of_open_complaints'], errors='coerce')
file.head(20)
```
